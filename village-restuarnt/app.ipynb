```python
import requests
from bs4 import BeautifulSoup

# Menu URL from the payload
menu_url = "https://www.villagesoulofindia.com/app/order/menu"

# Fetch the menu page
response = requests.get(menu_url)

# Check if the request is successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
   # Find all cards (menu items)
    cards = soup.find_all('div', class_='card')
    
    # Extract details for each card
    menu_items = []
    for card in cards:
        item = {}
        # Extract image URL
        img_tag = card.find('img', class_='card-img-top')
        item['image_url'] = img_tag['src'] if img_tag else None
    
        # Extract name
        name_tag = card.find('h5')
        item['name'] = name_tag.text.strip() if name_tag else None
    
        # Extract description
        desc_tag = card.find('span')
        item['description'] = desc_tag.text.strip() if desc_tag else None
    
        # Extract price
        price_tag = card.find('div', class_='text-muted')
        item['price'] = price_tag.text.strip() if price_tag else None
    
        # Add item to list
        menu_items.append(item)
    
        # Print all menu items
        for item in menu_items:
            print(f"Name: {item['name']}")
            print(f"Description: {item['description']}")
            print(f"Price: {item['price']}")
            print(f"Image URL: {item['image_url']}")
            print("-" * 40)
else:
    print(f"Failed to fetch menu. Status code: {response.status_code}")
```

```python
import requests

# Yelp API credentials
api_key = "9aNmlnpUIRRax9ldXhk0LRT_L5ZODAlZJTHAOj3xBtcRM8OXwuMQNI7eAU6yRu52q6o97PTXwTFADUU39tWbISCAmoudmGuuK0fX11UfaLejCG_JPL7UvP5Z5bJWZ3Yx"

headers = {"Authorization": f"Bearer {api_key}"}

# Search parameters
url = "https://api.yelp.com/v3/businesses/search"
params = {
    "latitude": 40.76656,  # Latitude of the reference restaurant
    "longitude": -73.52353,  # Longitude of the reference restaurant
    "radius": 2000,  # 2 km radius in meters
    "categories": "indian,vegetarian",  # Filter by Indian or vegetarian cuisine
    "sort_by": "rating",  # Sort by rating
    "limit": 6,  # Fetch up to 6 results
}

# Make the API request
response = requests.get(url, headers=headers, params=params)
top_alias_5 = []

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    restaurants = data.get("businesses", [])
    
    # Filter top 5 based on rating and print details
    top_5 = sorted(restaurants, key=lambda x: x['rating'], reverse=True)[1:6]
   
    for restaurant in top_5:
        top_alias_5.append(restaurant['alias'])
        print(f"Name: {restaurant['name']}")
        print(f"Rating: {restaurant['rating']}")
        print(f"Address: {', '.join(restaurant['location']['display_address'])}")
        print(f"Distance: {restaurant['distance']:.2f} meters")
        print(f"Menu URL: {restaurant.get('url', 'Not available')}")
        print("-" * 40)
else:
    print(f"Error: {response.status_code}, {response.text}")

print(top_alias_5) 
menu_details = {}

# Loop through top_alias_5
for alias in top_alias_5:
    url = f"https://api.yelp.com/v3/businesses/{alias}"
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        data = response.json()
        menu_details[alias] = data
        # print(json.dumps(data, indent=4))
    else:
        print(f"Error fetching details for {alias}: {response.status_code}")

print(json.dumps(menu_details, indent=4))

```
