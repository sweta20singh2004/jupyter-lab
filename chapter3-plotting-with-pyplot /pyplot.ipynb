Example 3: Tanushree is doing some research. She has a stored line of Pascal's triangle numbers as ar2 as shown below.

```python
import matplotlib.pyplot as pl 
import numpy as np 
ar2 = [1,7,21,35,35,21,7,1] 
s2 = np.sin(ar2) 
c2 = np.cos(ar2) 
t2 = np.tan(ar2) 
pl.plot(ar2,s2,'c') 
pl.plot(ar2, c2, 'r') 
pl.plot(ar2, t2, 'k', linestyle = 'dashed')  
pl.xlabel('Array values') 
pl.ylabel('Sine, Cosine and Tangent values') 
pl.show 
```

(o) The fibonacci series should be plotted as a cyan line with o markers having size as 5 and edge colour as red

Example 6: Create an array, A in the range 1 to 20 with values 1.25 apart. Another array, B, contains the log values of the elements in the array A.

### Adding Legends

```python
import matplotlib.pyplot as pl 
import numpy as np 
Val = [[5.,25.,45.,20.],[4.,23.,49.,17.],[6.,22.,47.,19.]] 
X = np.arange(4) 
pl.bar(X+0.00, Val[0], color = 'b', width = 0.25, label = 'range1') 
pl.bar(X+0.25, Val[1], color = 'c', width = 0.25, label = 'range2') 
pl.bar(X+0.50, Val[2], color = 'r', width = 0.25, label = 'range3') 
pl.legend(loc = 'upper left') 
pl.show() 
```

Example 31: Write a program to plot a histogram using prof. Awasthi's data and then draw a frequency polygon by joining the edges'midpoints through a line chart .

```python
import matplotlib.pyplot as pl 
import numpy as np 
import pandas as pd 
mu = 100 
sigma = 15 
x = mu + sigma*np.random.randn(100) 
pl.figure(figsize = (10,7)) 
y = np.arange(len(x)) 
n, edges, p = pl.hist(x, bins = 40, histtype = 'step') 
m = 0.5*(edges[1:]+edges[:-1]) 
m = m.tolist() 
l = len(m) 
m.insert(0,m[0]-10) 
m.append(m[l-1]+10) 
n = n.tolist() 
n.insert(0,0) 
n.append(0) 
pl.plot(m,n, '-^') 
pl.title('Research data Histogram') 
pl.show() 
fig1 = pl.gcf()
fig1.savefig('./research.png', dpi=100)
pl.show()
pl.draw()
```
