Plot the following data using line plot.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as pl 
Day = [1,2,3,4,5,6,7] 
Ticketssold = [2000,2800,3000,2500,2300,2500,1000] 
pl.title('Ticket collection') 
pl.plot(Day, Ticketssold, color = 'c') 
pl.xticks(Day, ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']) 
pl.xlabel('Days') 
pl.ylabel('Tickets Sold') 
pl.show() 
```
