Plot the following data using line plot.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as pl 
Day = [1,2,3,4,5,6,7] 
Ticketssold = [2000,2800,3000,2500,2300,2500,1000] 
pl.title('Ticket collection') 
pl.plot(Day, Ticketssold, color = 'c') 
pl.xticks(Day, ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']) 
pl.xlabel('Days') 
pl.ylabel('Tickets Sold') 
pl.show() 
```

2: Collect and store data related to the screen time of students in your class seperately for boys anf girls and present it using boxplot

```python
import matplotlib.pyplot as pl 
girls = [10.0,9.5,13.6,12,12,10,6,4,4,8,7.3,8.2] 
boys = [12.2,8.5,17.4,11.5,12.5,12,5,8.5,6,9.2,6.3,6.2] 
pl.boxplot([girls,boys], tick_labels = ['Girls','Boys']) 
pl.ylabel('Screen Time') 
pl.title('Teenagers Screen Time') 
pl.show() 
```
