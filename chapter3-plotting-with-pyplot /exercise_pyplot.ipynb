# Random

```python
import matplotlib.pyplot as pl 
import numpy as np 
x = np.random.randint(1,100,size = (250,))
y = np.random.randint(1,100,size = (250,))
z = np.random.randint(1,100,size = (250,))
a = np.random.randint(1,100,size = (250,))
b = np.random.randint(1,100,size = (250,))
sizes = np.random.randint(5, 60, size=250) 
pl.scatter(x,y ,color = 'r', s = sizes ) 
pl.scatter(x,z ,color = 'b', s = sizes ) 
pl.scatter(a,b ,color = 'g', s = sizes ) 
pl.scatter(b,x ,color = 'c', s = sizes ) 
pl.scatter(a,z ,color = 'y', s = sizes ) 
pl.xlabel("X Values")
pl.ylabel("Y Values") 
pl.savefig("randm.pdf")
pl.show() 
```

```python
import matplotlib.pyplot as pl
import numpy as np

# Create random data
x = np.random.randint(100, 250, size=50)  # 50 random x-values
y = np.random.randint(100, 250, size=50)  # 50 random y-values

# Define colors and sizes
info = ['r', 'y', 'c', 'k', 'g']
colors = np.random.choice(info, size=len(x))  # Randomly choose a color for each point
sizes = np.random.uniform(5, 10, size=len(x))  # Random sizes between 5 and 10

# Create scatter plot
pl.scatter(x, y, color=colors, s=sizes)

# Label the axes
pl.xlabel("X Values")
pl.ylabel("Y Values")

# Show the plot
pl.show()
```
