Plot the following data using line plot.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as pl 
Day = [1,2,3,4,5,6,7] 
Ticketssold = [2000,2800,3000,2500,2300,2500,1000] 
pl.title('Ticket collection') 
pl.plot(Day, Ticketssold, color = 'c') 
pl.xticks(Day, ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']) 
pl.xlabel('Days') 
pl.ylabel('Tickets Sold') 
pl.show() 
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as pl 
Week1 = [5000,6000,6500,3500,4000,5500,8000] 
Week2 = [4000,3000,5000,5500,3000,4500,5500] 
Week3 = [4000,5800,3500,2500,3000,5000,5900] 
c = (Week1, Week2, Week3) 
Day = [0,1,2,3,4,5,6] # This was the thing to be tweaked.
d = pd.DataFrame(c, index = [1,2,3]) 
# print(d.T) 
d.T.plot(kind = 'bar', width = 0.8) 
pl.legend(loc = 'upper left') 
pl.title('Mela Sales Report') 
pl.xticks(Day,['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
pl.xlabel('Days') 
pl.ylabel('Sales in Rs.') 
pl.show() 
```
