## Marks Visualizer

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Manually provided data
dates = ['2023-01-01', '2023-02-01', '2023-03-01']
physics = [85, 90, 78]
chemistry = [88, 85, 80]
maths = [95, 89, 92]
english = [75, 80, 78]
ip = [80, 82, 85]

# Creating a dataframe with the provided data
data = {
    'DATE': pd.to_datetime(dates),
    'PHYSICS': physics,
    'CHEMISTRY': chemistry,
    'MATHS': maths,
    'ENGLISH': english,
    'IP': ip
}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(14, 7))

plt.plot(df['DATE'], df['PHYSICS'], marker='o', label='PHYSICS')
plt.plot(df['DATE'], df['CHEMISTRY'], marker='o', label='CHEMISTRY')
plt.plot(df['DATE'], df['MATHS'], marker='o', label='MATHS')
plt.plot(df['DATE'], df['ENGLISH'], marker='o', label='ENGLISH')
plt.plot(df['DATE'], df['IP'], marker='o', label='IP')

plt.xlabel('Date')
plt.ylabel('Marks')
plt.title('Marks in Different Subjects Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
